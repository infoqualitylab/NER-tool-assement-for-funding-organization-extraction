# Combining all the extracted acknowledgement statements

**Importing required Libraries**

```python
import pandas as pd
import numpy as np
import os
```

**Input path**

```python
input_path = "../data/ack_data"
ack_files = os.listdir(input_path)
ack_files = [ele for ele in ack_files if ele.split(".")[-1] == 'csv']
```

```python
cumulative_data = pd.DataFrame()
file_count = 0
for ele in ack_files:
    print(ele)
    data = pd.read_csv(os.path.join(input_path, ele), encoding='iso-8859-1')
    # data = data.dropna(subset=['ACK', 'ACKNOWLEDGEMENT', 'FUNDING', 'FOOTNOTES'], how='all')
    data = data.replace(np.nan, '')
    cumulative_data = pd.concat([cumulative_data, data])
    file_count+=1
```

```python
l = []
cumulative_data["Funding Statement"] = cumulative_data['ACK'].astype(str) + " " + cumulative_data['ACKNOWLEDGEMENT'].astype(str) + " " + cumulative_data['FUNDING'].astype(str) + " " + cumulative_data['FOOTNOTES'].astype(str)
cumulative_data = cumulative_data.replace('   ', 'NANANANA')
cumulative_data['Funding Statement'] = cumulative_data['Funding Statement'].str.strip(" ")
for ele in cumulative_data['Funding Statement']:
    ele = ele.split(" ")
    ele = [e.strip() for e in ele]
    ele = " ".join(ele)
    l.append(ele)
cumulative_data['acknowledgement'] = l
cumulative_data = cumulative_data[['Article_Title', 'PMC_ID', 'DOI', 'acknowledgement']]
```
